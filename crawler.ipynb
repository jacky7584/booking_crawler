```python
from selenium import webdriver
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

proxy = "10.11.1.78:8080"
proxies = {
    "http": "http://" + proxy,
    "https": "http://" + proxy,
}

options = Options()
options.add_argument("start-maximized")
options.page_load_strategy = 'eager'
options.add_argument("headless")  # 在后台运行，无界面模式

# 创建一个Edge浏览器实例，传递浏览器选项
edge = webdriver.Edge(options=options)

# 要爬取的网页URL
url = "https://aomp109.judicial.gov.tw"  # 替换为目标动态网站的URL
#driver = webdriver.Edge()
# 打开网页
edge.get(url)

# 设置等待时间
wait = WebDriverWait(edge, 10)  # 最多等待10秒
page_source = driver.page_source
print(page_source)
try:
    # 使用 By.CLASS_NAME 和 By.ID 定位元素
    button = edge.find_element(By.CLASS_NAME, "btn-danger")
    if button.get_attribute("id") == "btn_ok":
        button.click()
        print("按钮已点击")
    else:
        print("找到的按钮不是目标按钮")

finally:
    # 关闭浏览器
    driver.quit()



# 打印网页源代码
print(page_source)

# 关闭浏览器
driver.quit()
```
